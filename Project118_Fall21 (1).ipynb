```python
sum_pop = sum(FA["Pop2010"])
sum_pop

low_access = sum(FA[FA["Urban"] == 1]["lapop1"].dropna()) + sum(FA[FA["Urban"] == 0]["lapop10"].dropna())
low_access

prop_la = low_access/sum_pop
print(prop_la)
```

```python
FA["prop_la_white1"] = FA[FA["Urban"] == 1]["lawhite1"]/ FA["TractWhite"]
FA["prop_la_black1"] = FA[FA["Urban"] == 1]["lablack1"]/ FA["TractBlack"]
FA["prop_la_asian1"] = FA[FA["Urban"] == 1]["laasian1"]/ FA["TractAsian"]
FA["prop_la_hisp1"] = FA[FA["Urban"] == 1]["lahisp1"]/ FA["TractHispanic"]
FA

##rural analysis

FA["prop_la_white10"] = FA[FA["Urban"] == 0]["lawhite10"]/ FA["TractWhite"]
FA["prop_la_black10"] = FA[FA["Urban"] == 0]["lablack10"]/ FA["TractBlack"]
FA["prop_la_asian10"] = FA[FA["Urban"] == 0]["laasian10"]/ FA["TractAsian"]
FA["prop_la_hisp10"] = FA[FA["Urban"] == 0]["lahisp10"]/ FA["TractHispanic"]



FA


white_la_urban = sum(FA[FA["Urban"] == 1]["lawhite1"].dropna())/ sum(FA[FA["Urban"] == 1]["TractWhite"].dropna())
black_la_urban = sum(FA[FA["Urban"] == 1]["lablack1"].dropna())/ sum(FA[FA["Urban"] == 1]["TractBlack"].dropna())
asian_la_urban = sum(FA[FA["Urban"] == 1]["laasian1"].dropna())/ sum(FA[FA["Urban"] == 1]["TractAsian"].dropna())
hispanic_la_urban = sum(FA[FA["Urban"] == 1]["lahisp1"].dropna())/ sum(FA[FA["Urban"] == 1]["TractHispanic"].dropna())
hispanic_la_urban = sum(FA[FA["Urban"] == 1]["lahisp1"].dropna())/ sum(FA[FA["Urban"] == 1]["TractHispanic"].dropna())
hopi_la_urban = sum(FA[FA["Urban"] == 1]["lanhopi1"].dropna())/ sum(FA[FA["Urban"] == 1]["TractNHOPI"].dropna())
aian_la_urban = sum(FA[FA["Urban"] == 1]["laaian1"].dropna())/ sum(FA[FA["Urban"] == 1]["TractAIAN"].dropna())
omultir_la_urban = sum(FA[FA["Urban"] == 1]["laomultir1"].dropna())/ sum(FA[FA["Urban"] == 1]["TractOMultir"].dropna())
la_urban = sum(FA[FA["Urban"] == 1]["lapop1"].dropna())/ sum(FA[FA["Urban"] == 1]["Pop2010"].dropna())


white_la_rural = sum(FA[FA["Urban"] == 0]["lawhite10"].dropna())/ sum(FA[FA["Urban"] == 0]["TractWhite"].dropna())
black_la_rural = sum(FA[FA["Urban"] == 0]["lablack10"].dropna())/ sum(FA[FA["Urban"] == 0]["TractBlack"].dropna())
asian_la_rural = sum(FA[FA["Urban"] == 0]["laasian10"].dropna())/ sum(FA[FA["Urban"] == 0]["TractAsian"].dropna())
hispanic_la_rural = sum(FA[FA["Urban"] == 0]["lahisp10"].dropna())/ sum(FA[FA["Urban"] == 0]["TractHispanic"].dropna())
hopi_la_rural = sum(FA[FA["Urban"] == 0]["lanhopi10"].dropna())/ sum(FA[FA["Urban"] == 0]["TractNHOPI"].dropna())
aian_la_rural = sum(FA[FA["Urban"] == 0]["laaian10"].dropna())/ sum(FA[FA["Urban"] == 0]["TractAIAN"].dropna())
omultir_la_rural = sum(FA[FA["Urban"] == 0]["laomultir10"].dropna())/ sum(FA[FA["Urban"] == 0]["TractOMultir"].dropna())

kids_la_urban = sum(FA[FA["Urban"] == 1]["lakids1"].dropna())/ sum(FA[FA["Urban"] == 1]["TractKids"].dropna())
kids_la_rural = sum(FA[FA["Urban"] == 0]["lakids10"].dropna())/ sum(FA[FA["Urban"] == 0]["TractKids"].dropna())
print(kids_la_urban)
print(kids_la_rural)
seniors_la_urban = sum(FA[FA["Urban"] == 1]["laseniors1"].dropna())/ sum(FA[FA["Urban"] == 1]["TractSeniors"].dropna())
seniors_la_rural = sum(FA[FA["Urban"] == 0]["laseniors10"].dropna())/ sum(FA[FA["Urban"] == 0]["TractSeniors"].dropna())
print(seniors_la_urban)
print(seniors_la_rural)


la_urban = sum(FA[FA["Urban"] == 1]["lapop1"].dropna())/ sum(FA[FA["Urban"] == 1]["Pop2010"].dropna())
print(la_urban)
la_rural = sum(FA[FA["Urban"] == 0]["lapop10"].dropna())/ sum(FA[FA["Urban"] == 0]["Pop2010"].dropna())
print(la_rural)


```

```python
FA_urban = FA[FA["Urban"] == 1]

FA_state_group = pd.DataFrame(FA_urban.groupby("State")["Pop2010", "PovertyRate","MedianFamilyIncome","lapop1", "lapop10","lakids1","laseniors1", "lawhite1", "lablack1","laasian1","lahisp1","TractKids", "TractSeniors", "TractWhite","TractBlack","TractAsian","TractHispanic"].sum())
FA_state_group["prop_la_urban"] = FA_state_group["lapop1"] / FA_state_group["Pop2010"]
FA_state_group["prop_la_kids"] = FA_state_group["lakids1"] / FA_state_group["TractKids"]
FA_state_group["prop_la_seniors"] = FA_state_group["laseniors1"] / FA_state_group["TractSeniors"]
FA_state_group["prop_la_white"] = FA_state_group["lawhite1"] / FA_state_group["TractWhite"]
FA_state_group["prop_la_black"] = FA_state_group["lablack1"] / FA_state_group["TractBlack"]
FA_state_group["prop_la_asian"] = FA_state_group["laasian1"] / FA_state_group["TractAsian"]
FA_state_group["prop_la_hispanic"] = FA_state_group["lahisp1"] / FA_state_group["TractHispanic"]
FA_state_group['State_'] = FA_state_group.index
FA_state_group.head(5)

la_urban = sum(FA_state_group["lapop1"])/sum(FA_state_group["Pop2010"])
la_urban

FA_state_group["prop_la_urban"].sort_values()
#print(np.mean(FA_state_group["prop_la_kids"]))
#print(np.mean(FA_state_group["prop_la_seniors"]))
#state_group = state_group.sort_values(by=["low_access"], ascending=True)

```

```python
FA_rural = FA[FA["Urban"] == 0]
FA_state_group_rural = pd.DataFrame(FA_rural.groupby("State")["Pop2010", "lapop10"].sum())
FA_state_group_rural["low_access_prop"] = FA_state_group_rural["lapop10"] / FA_state_group_rural["Pop2010"]
FA_state_group_rural['State_'] = FA_state_group_rural.index
FA_state_group_rural.head(5)
la_rural = sum(FA_state_group_rural["lapop10"])/sum(FA_state_group_rural["Pop2010"])
la_rural

FA_state_group_rural["low_access_prop"].sort_values()
```

```python
plt.rcParams["figure.figsize"] = (20,15)
plt.barh(FA_state_group['State_'], FA_state_group["prop_la_urban"], label = "urban")
plt.barh(FA_state_group_rural['State_'], FA_state_group_rural["low_access_prop"], label = "rural")
plt.xlabel("Prop of population with low access")
plt.ylabel("State")
plt.title("proportion of population low food access in each state")
plt.legend(loc='upper right')

```
